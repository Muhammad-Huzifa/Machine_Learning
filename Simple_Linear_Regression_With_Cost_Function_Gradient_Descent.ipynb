```python
class NewML:
    def __init__(self):
        self.m = None  # Slope
        self.b = None  # Intercept

    def fit(self, X, y):
        num = 0
        den = 0
        for i in range(len(X)):
            num += (X[i] - X.mean()) * (y[i] - y.mean())
            den += (X[i] - X.mean()) ** 2
        self.m = num / den
        self.b = y.mean() - (self.m * X.mean())
        print(self.m)
        print(self.b)

    def predict(self, X):
        return self.m * X + self.b

lr = NewML()
lr.fit(X, y)
y_pred = lr.predict(X)
print(y_pred)
plt.scatter(X, y, label='Actual Values')
plt.scatter(X, y_pred, c='r', label='Predicted')
plt.plot(X, y_pred, c='g')
plt.figure(figsize=(1, 1))
```
