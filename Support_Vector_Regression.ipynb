```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("C:\\Users\\PMLS\\Desktop\\Data_Sets\\Salary_Data.csv")
df
X =  df['YearsExperience'].values
y = df['Salary'].values
print(X, y)
X = np.reshape(X,(-1,1))
```

### SUPPORT_VECTOR_REGRESSION

#### SVR_WITH_KERNEL_'PLOY'

```python
x_Scalar = StandardScaler()
y_Scalar = StandardScaler()
X_Scaled = x_Scalar.fit_transform(X)
y_Scaled = y_Scalar.fit_transform(np.reshape(y,(-1,1)))
model_S = SVR(kernel = 'poly' ,C =2 )
model_S.fit(X_Scaled ,y_Scaled[:,0])
```

```python
y_temp = model_S.predict(X_Scaled)
y_predicted = y_Scalar.inverse_transform(y_temp.reshape(-1,1))
y_predicted[:,0]
```

#### To Plot the  Model Values(Using MatplotLib)

```python
plt.scatter(X,y,label = 'Actual_Values' , c = 'g')
plt.plot(X,y_predicted,label = 'MOdel',c = 'b' )
plt.scatter(X,y_predicted,label = 'Predicted_Values',c = 'r')
plt.legend()
plt.xlabel('Years_Of_Experince')
plt.ylabel('Salary_Per_Month')
plt.xticks(range(1,13,1))
plt.yticks(range(30000,135000,10000),)
plt.title('Support_Vector_Regression_With_Kernel_Poly')
plt.grid(True)
Score_SVR_Poly = model_S.score
Score_SVR_Poly(X_Scaled,y_Scaled)
```

#### SVR_WITH_KERNEL_'rbf'

```python
x_Scalar = StandardScaler()
y_Scalar = StandardScaler()
X_Scaled = x_Scalar.fit_transform(X)
y_Scaled = y_Scalar.fit_transform(np.reshape(y,(-1,1)))
model_S = SVR(kernel = 'rbf' ,C =2 )
model_S.fit(X_Scaled ,y_Scaled[:,0])
y_temp = model_S.predict(X_Scaled)
y_predicted = y_Scalar.inverse_transform(y_temp.reshape(-1,1))
y_predicted[:,0]
```

```python
plt.scatter(X,y,label = 'Actual_Values' , c = 'g')
plt.plot(X,y_predicted,label = 'Model',c = 'b' )
plt.scatter(X,y_predicted,label = 'Predicted_Values',c = 'r')
plt.legend()
plt.xlabel('Years_Of_Experince')
plt.ylabel('Salary_Per_Month')
plt.xticks(range(1,13,1))
plt.yticks(range(30000,135000,10000),)
plt.title('Support_Vector_Regression_With_Kernel_rbf')
plt.grid(True)
Score_SVR_rbf = model_S.score
Score_SVR_rbf(X_Scaled,y_Scaled)
```

### SVR_WITH_KERNEL_'Linear'

```python
x_Scalar = StandardScaler()
y_Scalar = StandardScaler()
X_Scaled = x_Scalar.fit_transform(X)
y_Scaled = y_Scalar.fit_transform(np.reshape(y,(-1,1)))
model_S = SVR(kernel = 'linear' ,C =2 )
model_S.fit(X_Scaled ,y_Scaled[:,0])
y_temp = model_S.predict(X_Scaled)
y_predicted = y_Scalar.inverse_transform(y_temp.reshape(-1,1))
y_predicted[:,0]
```

```python
plt.scatter(X,y,label = 'Actual_Values' , c = 'g')
plt.plot(X,y_predicted,label = 'Model',c = 'b' )
plt.scatter(X,y_predicted,label = 'Predicted_Values',c = 'r')
plt.legend()
plt.xlabel('Years_Of_Experince')
plt.ylabel('Salary_Per_Month')
plt.xticks(range(1,13,1))
plt.yticks(range(30000,135000,10000),)
plt.title('Support_Vector_Regression_With_Kernel_linear')
plt.grid(True)
Score_SVR_lin = model_S.score
Score_SVR_lin(X_Scaled,y_Scaled)
```

```python
print(f'The Score of Support Vector Regression is with Kernel "Poly" is {Score_SVR_Poly(X_Scaled,y_Scaled)}')
print(f'The Score of Support Vector Regression is with Kernel "rbf" is {Score_SVR_rbf(X_Scaled,y_Scaled)}')
print(f'The Score of Support Vector Regression is with Kernel "linear" is {Score_SVR_lin(X_Scaled,y_Scaled)}')

```
