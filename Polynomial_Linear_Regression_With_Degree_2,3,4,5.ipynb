```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## DataSet Reperesentation

```python
df = pd.read_csv("C:\\Users\\PMLS\\Desktop\\Data_Sets\\Salary_Data.csv")
df
```

```python
X = df['YearsExperience'].values
y = df['Salary'].values
X,y
```

```python
X = np.reshape(X,(-1,1))
X
```

## Ploynomial_Linear_Regression With Degree 2,3,4,5Â¶

### Ploynomial_Linear_Regression With Degree 2

```python
pt = PolynomialFeatures(degree = 2)
X_poly = pt.fit_transform(X)
X , X_poly
```

```python
model2 = linear_model.LinearRegression()
model2.fit(X_poly , y)
```

```python
ypredicted2 = model2.predict(X_poly)
ypredicted2 
```

```python
plt.scatter(X  , y  , label = 'Actual_Values' ,c = 'b')
plt.scatter(X , ypredicted2 , label = 'Predicted_Values' , c = 'r')
plt.plot(X, ypredicted2 , label = 'Model')
plt.xlabel('Years_Of_Experince')
plt.ylabel('Salary_Per_Month')
plt.xticks(range(1,13,1))
plt.yticks(range(30000,135000,10000))
plt.title('Ploynomial_Regression_With_Degree_2')
plt.grid(True)
```

```python
Score_PLR_2=model2.score(X_poly,y)
Score_PLR_2
```
